```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
```

#Pie chart showing various components of total energy for He atom

```python
energies = [-2.834836, 2.767924, 1.996121, -6.625567, -0.973314]
labels = ['E_tot', 'E_kin', 'E_coul', 'E_enuc', 'E_xc']

# Calculate the absolute value of the energies
abs_energies = [abs(energy) for energy in energies]

# Calculate the percentage of each energy for the pie chart
percentages = [energy / sum(abs_energies) * 100 for energy in abs_energies]

# Create the pie chart
fig, ax = plt.subplots()
ax.pie(abs_energies, labels=labels, autopct='%1.1f%%', startangle=140)

# Set the aspect ratio to be equal, so the pie chart is circular
ax.axis('equal')

# Display the pie chart
plt.show()
```

#Graph showing trend of Total Energies as function of atomic number

```python
atomic_numbers = np.arange(1, 13)

# The total energies for each element from Hydrogen to Magnesium
total_energies = [
    -0.445671, -2.834836, -7.335198, -14.447212, -24.344201,
    -37.425753, -54.025018, -74.473082, -99.099654, -128.233477,
    -161.440062, -199.139581
]

# Element symbols for the plot
element_symbols = [
    'H', 'He', 'Li', 'Be', 'B', 'C', 'N', 'O', 'F', 'Ne', 'Na', 'Mg'
]

# Set the style for the plot
sns.set_theme()
total_energies=-np.array(total_energies)
# Create the plot
plt.plot(atomic_numbers, total_energies, label='Calculated Total Energies', marker='o')

# Annotate each point with the corresponding element symbol
for z, energy, symbol in zip(atomic_numbers, total_energies, element_symbols):
    plt.text(z, energy, symbol, ha='center', va='bottom')

plt.yscale('log')
# Add labels and title
plt.xlabel('Z (Atomic Number)')
plt.ylabel('Total Energy (eV)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

#Graph showing trend of Ionisation Energies as function of atomic number

```python
# Ionization energies in eV for elements from Hydrogen (Z=1) to Magnesium (Z=12)
calculated_energies = [12.122251, 26.481866, 5.232654, 9.506971, 8.321051,
                        10.985998, 13.734096, 16.590368, 19.566565, 22.668670,
                        5.160520, 8.051554]
experimental_energies = [13.598, 24.587, 5.392, 9.323, 8.298,
                         11.260, 14.534, 13.618, 17.423, 21.565,
                         5.139, 7.646]

# Atomic numbers corresponding to elements from Hydrogen (Z=1) to Magnesium (Z=12)
atomic_numbers = np.arange(1, 13)

# Symbols for elements from Hydrogen to Magnesium
symbols = ['H', 'He', 'Li', 'Be', 'B', 'C', 'N', 'O', 'F', 'Ne', 'Na', 'Mg']

# Create a figure and a set of subplots
plt.figure()

# Plotting the calculated ionization energies
plt.plot(atomic_numbers, calculated_energies, label='DFT-LDA', marker='.', color='tab:blue')

# Adding the scatter points for experimental ionization energies
plt.scatter(atomic_numbers, experimental_energies, label='Experimental', color='tab:orange')

# Annotate each point with the element symbol
for i, symbol in enumerate(symbols):

    plt.annotate(symbol, (atomic_numbers[i], experimental_energies[i]),
                 textcoords="offset points", xytext=(0,10), ha='center')


# Adding labels and title
plt.xlabel('Z (Atomic Number)')
plt.ylabel('Ionisation Energy (eV)')

# Adding legend
plt.legend()

# Show the plot
plt.show()
```

#Graph showing trend of Energy of orbitals as function of atomic number

```python
# Define the atomic numbers for Hydrogen to Magnesium (Z = 1 to 12)
atomic_numbers = np.arange(1, 13)

# The energy levels for each orbital (1s, 2s, 2p, 3s)
orbital_energies = {
    '1s': [-0.233471, -0.570424, -1.878563, -3.85641, -6.564346, -9.947717, -14.011501, -18.758238, -24.189413, -30.30634, -37.719969, -45.973245],
    '2s': [np.nan, np.nan, -0.105539, -0.205743, -0.344699, -0.500863, -0.676148, -0.871356, -1.086863, -1.323003, -2.063393, -2.903753],
    '2p': [np.nan, np.nan, np.nan, np.nan, -0.136601, -0.199183, -0.266294, -0.338375, -0.41561, -0.498226, -1.060629, -1.718974],
    '3s': [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, -0.103413, -0.175428]
}

# Set the seaborn theme for the plot
sns.set_theme()


# Plot each orbital energy trend
for orbital, energies in orbital_energies.items():
    valid_indices = ~np.isnan(energies)
    plt.plot(atomic_numbers[valid_indices], -np.array(energies)[valid_indices], label=f'{orbital} orbital', marker='o')

plt.yscale('log')
# Labeling the axes
plt.xlabel('Z (Atomic Number)')
plt.ylabel('Orbital Energy (eV)')
plt.legend()

# Show the plot with a grid
plt.grid(True)
plt.show()
```
